```python
# parrent class
class character:
  """Bantuan Library karakter:

  NAME
    Karakter

  DESCRIPTION
    Libary unruk belajar bermain game RPG sederhana menggunakan Python
    ==================================================================
    Library yang membantu para player pemula untuk bermain game dan 
    developer untuk mengembangkan karakter
  
  CLASES
    witcher
    knight

  FUNCTION
    get_info_character() -> tuple
    menyerang() -> int
    bertahan() -> int
    terkena_serangan() -> int
    heal() -> int
    run() -> int
    up_attack() -> int
    up_block() -> int
    recover_potion() -> int
    menyerang_penuh() -> int
    kemampuan() -> str
  """

  nama = ''
  health = 0
  stamina = 0
  block = 0
  attack = 0
  potion = 0

  def get_info_character(self):
    return self.nama, self.health, self.stamina, self.potion, self.skill, self.block, self.attack

  def menyerang(self):
    if self.potion > 10:
      print(f'{self.nama} menyerang dengan {self.skill} sebesar {self.attack}')
      return self.potion - 10
    if self.potion <= 0:
      print("potionmu kurang")
      return 0

  def recover_potion(self):
    print(f'{self.nama} melakukan recovery potion')
    return self.potion + 10

  def heal(self):
    print(f'{self.nama} sedang melakukan healing')
    return self.health + 10

  def up_block(self):
    print(f'{self.nama} memakai potion menambah blocking')
    return self.block + 10

  def up_attack(self):
    print(f'{self.nama} memakai potion menambah attacking')
    return self.attack + 10

  def run(self):
    if self.stamina > 0:
      print(f'{self.nama} sedang berlari')
    if self.stamina <= 0:
      print("stamina tidak cukup")
      return 0

  def bertahan(self, serangan):
    if serangan > self.block:
      print(f'{self.nama} terkena serangan menggunakan {self.skill} sebesar {serangan - self.block}')
      self.health = self.health - (serangan - self.block)
      print(f'health {self.nama} adalah {self.health}')
    if serangan == self.block:
      print(f'serangan tidak berdampak pada musuh')
    if self.hp == 0:
      print(f'{self.nama} menunggu respawn')

class witcher(character):
  stick = ''

  def serangan_penuh(self):
    if self.potion > 50:
      print(f'{self.nama} mengeluarkan skill pamungkas')
      return self.attack * 10 and self.potion - 10
    else:
      print("potion kurang")
      return 0

class knight(character):
  sword = ''

  def serangan_penuh(self):
    if self.potion > 10:
      print(f'{self.nama} mengeluarkan skill pamungkas')
      return self.attack * 10 and self.potion - 10
    else:
      print("potion kurang")
      return 0
```
